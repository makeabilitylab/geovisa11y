```python
import pandas as pd
import geopandas as gpd
```

```python
states = gpd.read_file('data/00_state/state.shp')
```

```python
# Perform a spatial join to find neighboring states
neighbors_df = gpd.sjoin(states, states, how="left", predicate="touches")
```

```python
neighbors_df[['state_name_right','state_name_left']]
```

```python
neighbors_dict = (
    neighbors_df.groupby("state_name_right")["state_name_left"]
    .apply(lambda x: list(x.unique()))
    .to_dict()
)

# Map the dictionary to the states GeoDataFrame
states["a_neighbor"] = states["state_name"].map(neighbors_dict).fillna("").apply(lambda x: [] if x == "" else x)
```

```python
#save to shapefile
states.to_file('data/00_state/state.shp')
```
